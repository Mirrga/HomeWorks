```python
def check_balance(s):
    open_count = s.count('(')
    close_count = s.count(')')
    return open_count == close_count

str1 = "((a + b - c) * (c - d + a))"
if check_balance(str1):
    print("Количество открывающихся и закрывающихся скобок одинаково.")
else:
    print("Количество открывающихся и закрывающихся скобок не одинаково.")
```

```python
import random

def fill_file(file, count, lower_bound, upper_bound):
    with open(file, 'w') as f:
        for _ in range(count):
            number = random.randint(lower_bound, upper_bound)
            f.write(f"{number}\n")

def filter(input_file, output_file, m, n):
    with open(input_file, 'r') as f:
        numbers = f.readlines()
    with open(output_file, 'w') as g:
        for number in numbers:
            num = int(number.strip())
            if num % m == 0 and num % n != 0:
                g.write(f"{num}\n")

file_f = 'f.txt'
file_g = 'g.txt'
count_of_numbers = 100
lower_bound = 1
upper_bound = 100
m = 3
n = 5
fill_file(file_f, count_of_numbers, lower_bound, upper_bound)
filter(file_f, file_g, m, n)
print(f"Файл '{file_f}' заполнен случайными числами.")
print(f"Файл '{file_g}' содержит числа, которые делятся на {m} и не делятся на {n}.")
```

```python
def check_baggage(file):
    with open(file, 'r') as file:
        for line in file:
            count, weight = line.strip().split()
            count = int(count)
            weight = float(weight)
            if count == 1 and weight < 30:
                return True
    return False

file = 'Bagazh.txt'
if check_baggage(file):
    print("Существует пассажир с одной вещью весом менее 30 кг.")
else:
    print("Пассажиров с одной вещью весом менее 30 кг не найдено.")
```

```python
def count_in_file(file, word):
    with open(file, 'r', encoding='utf-8') as file:
        text = file.read()
        count = text.count(word)
        return count

file = 'text.txt'
word = input("Введите слово для поиска: ")
count = count_in_file(file, word)

if count > 0:
    print(f"Слово '{word}' встречается {count} раз(а) в тексте.")
else:
    print(f"Слово '{word}' не найдено в тексте.")
```
