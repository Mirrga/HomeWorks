```python
def word_frequency(text):
    words = text.split()
    freq = {}
    for word in words:
        freq[word] = freq.get(word, 0) + 1
    return freq
```

```python
def invert_dict(d):
    inverted = {}
    for k, v in d.items():
        if v in inverted:
            if isinstance(inverted[v], list):
                inverted[v].append(k)
            else:
                inverted[v] = [inverted[v], k]
        else:
            inverted[v] = k
    return inverted
```

```python
def merge_dicts(d1, d2):
    merged = d1.copy()
    for k, v in d2.items():
        merged[k] = merged.get(k, 0) + v
    return merged
```

```python
def group_by_key_length(d):
    grouped = {}
    for k in d:
        length = len(k)
        if length not in grouped:
            grouped[length] = [k]
        else:
            grouped[length].append(k)
    return grouped
```

```python
def combine_values(d1, d2):
    combined = {}
    for k in set(d1) | set(d2):
        values = []
        if k in d1:
            values.append(d1[k])
        if k in d2:
            values.append(d2[k])
        combined[k] = values
    return combined
```
