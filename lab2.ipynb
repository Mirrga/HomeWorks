```python
class Product:
    def __init__(self, name, price, quantity):
        self.__name = name
        self.price = price
        self.quantity = quantity

    def total_value(self):
        return self.quantity * self.price

    def update_price(self, new_price):
        self.price = new_price

class CartItem(Product):
    def __init__(self, name, price, cart_quantity):
        super().__init__(name, price, 0)
        self.cart_quantity = cart_quantity
    def update_quantity(self, new_quantity):
        self.cart_quantity = new_quantity
    def total_value(self):
        return self.cart_quantity * self.price

product = CartItem("мыло", 100, 5)
print(product.total_value())
product.update_quantity(15)
print(product.total_value())
```

```python
class Shelf:
    def __init__(self, cabinet):
        self.cabinet = cabinet
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def count_items(self):
        return len(self.items)

class Cabinet:
    def __init__(self):
        self.shelves = []

    def add_shelf(self):
        new_shelf = Shelf(self)
        self.shelves.append(new_shelf)

    def count_shelves(self):
        return len(self.shelves)

cabinet = Cabinet()

cabinet.add_shelf()
cabinet.add_shelf()

print(f"Количество полок в шкафу: {cabinet.count_shelves()}")

cabinet.shelves[0].add_item("Книга1")
cabinet.shelves[0].add_item("Книга2")
cabinet.shelves[1].add_item("Книга")

print(f"Количество предметов на первой полке: {cabinet.shelves[0].count_items()}")
print(f"Количество предметов на второй полке: {cabinet.shelves[1].count_items()}")
```

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def __str__(self):
        return f'"{self.title}" by {self.author}'


class Reader:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return self.name


class Library:
    def __init__(self):
        self.books = []
        self.readers = []

    def add_book(self, book):
        self.books.append(book)

    def add_reader(self, reader):
        self.readers.append(reader)

    def list_books(self):
        return [str(book) for book in self.books]

    def list_readers(self):
        return [str(reader) for reader in self.readers]



library = Library()
book1 = Book("1984", "George Orwell")
book2 = Book("To Kill a Mockingbird", "Harper Lee")

library.add_book(book1)
library.add_book(book2)

reader1 = Reader("Alice")
reader2 = Reader("Bob")

library.add_reader(reader1)
library.add_reader(reader2)

print("Список книг в библиотеке:")
for book in library.list_books():
    print(book)

print("\nСписок читателей в библиотеке:")
for reader in library.list_readers():
    print(reader)
```

```python
from abc import ABC, abstractmethod


class TwoDShape(ABC):
    @abstractmethod
    def calculatePerimeter(self):
        pass

class Square(TwoDShape):
    def __init__(self, side):
        self.side = side

    def calculatePerimeter(self):
        return 4 * self.side

class Triangle(TwoDShape):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def calculatePerimeter(self):
        return self.a + self.b + self.c

square = Square(5)
triangle = Triangle(3, 4, 5)

print(square.calculatePerimeter())
print(triangle.calculatePerimeter())
```

```python
from abc import ABC, abstractmethod

class IProduct(ABC):
    @abstractmethod
    def getTotalValue(self):
        pass

    @abstractmethod
    def updatePrice(self, price: float):
        pass

class Product(IProduct):
    def __init__(self, name: str, price: float, quantity: int):
        self.name = name
        self.price = price
        self.quantity = quantity

    def getTotalValue(self):
        return self.price * self.quantity

    def updatePrice(self, price: float):
        self.price = price

product = Product("Laptop", 1000.0, 5)
print(product.getTotalValue())
product.updatePrice(1200.0)
print(product.getTotalValue())
```

```python
from abc import ABC, abstractmethod

class CartItem(ABC):
    def __init__(self, price: float, quantity: int):
        self.price = price
        self.quantity = quantity

    @abstractmethod
    def calculateTotalPrice(self):
        pass

class DiscountedItem(CartItem):
    def __init__(self, price: float, quantity: int, discount: float):
        super().__init__(price, quantity)
        self.discount = discount

    def calculateTotalPrice(self):
        return (self.price * self.quantity) * (1 - self.discount)

class PremiumItem(CartItem):
    def __init__(self, price: float, quantity: int, premium_rate: float):
        super().__init__(price, quantity)
        self.premium_rate = premium_rate

    def calculateTotalPrice(self):
        return (self.price * self.quantity) * (1 + self.premium_rate)

discounted_item = DiscountedItem(1000.0, 2, 0.1)
premium_item = PremiumItem(2000.0, 3, 0.2)

print(discounted_item.calculateTotalPrice())
print(premium_item.calculateTotalPrice())
```
